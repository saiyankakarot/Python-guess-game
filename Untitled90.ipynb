```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, Flatten, Reshape
from tensorflow.keras.losses import binary_crossentropy

# Load MNIST dataset
(x_train, _), (x_test, _) = mnist.load_data()
x_train = x_train.astype('float32') / 255.
x_test = x_test.astype('float32') / 255.
x_train = np.reshape(x_train, (len(x_train), 28, 28, 1))
x_test = np.reshape(x_test, (len(x_test), 28, 28, 1))
```

```python
input_img = Input(shape=(28, 28, 1))
x = Flatten()(input_img)
encoded = Dense(64, activation='relu')(x)
decoded = Dense(784, activation='sigmoid')(encoded)
decoded = Reshape((28, 28, 1))(decoded)

vanilla_autoencoder = Model(input_img, decoded)
vanilla_autoencoder.compile(optimizer='adam', loss='binary_crossentropy')

print("Training Vanilla Autoencoder...")
vanilla_autoencoder.fit(x_train, x_train, epochs=5, batch_size=256, validation_data=(x_test, x_test), verbose=1)

decoded_imgs_vanilla = vanilla_autoencoder.predict(x_test[:5])
```

```python
noise_factor = 0.3
x_train_noisy = np.clip(x_train + noise_factor * np.random.normal(loc=0.0, scale=1.0, size=x_train.shape), 0., 1.)
x_test_noisy = np.clip(x_test + noise_factor * np.random.normal(loc=0.0, scale=1.0, size=x_test.shape), 0., 1.)

input_noisy = Input(shape=(28, 28, 1))
x = Flatten()(input_noisy)
encoded = Dense(64, activation='relu')(x)
decoded = Dense(784, activation='sigmoid')(encoded)
decoded = Reshape((28, 28, 1))(decoded)

denoising_autoencoder = Model(input_noisy, decoded)
denoising_autoencoder.compile(optimizer='adam', loss='binary_crossentropy')

print("Training Denoising Autoencoder...")
denoising_autoencoder.fit(x_train_noisy, x_train, epochs=5, batch_size=256, validation_data=(x_test_noisy, x_test), verbose=1)

decoded_imgs_denoising = denoising_autoencoder.predict(x_test_noisy[:5])
```

```python
n = 5
plt.figure(figsize=(12, 6))

for i in range(n):
    # Original
    ax = plt.subplot(3, n, i + 1)
    plt.imshow(x_test[i].reshape(28, 28), cmap="gray")
    plt.title("Original")
    plt.axis("off")

    # Vanilla Output
    ax = plt.subplot(3, n, i + 1 + n)
    plt.imshow(decoded_imgs_vanilla[i].reshape(28, 28), cmap="gray")
    plt.title("Vanilla AE")
    plt.axis("off")

    # Denoising Output
    ax = plt.subplot(3, n, i + 1 + 2 * n)
    plt.imshow(decoded_imgs_denoising[i].reshape(28, 28), cmap="gray")
    plt.title("Denoising AE")
    plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
import tensorflow as tf
from tensorflow.keras import datasets, layers, models
import matplotlib.pyplot as plt

(train_images, train_labels), (test_images, test_labels) = datasets.mnist.load_data()

train_images = train_images / 255.0
test_images = test_images / 255.0


model = models.Sequential([
    layers.Flatten(input_shape=(28, 28)),
    layers.Dense(128, activation='relu'),
    layers.Dense(10, activation='softmax')
])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model.fit(train_images, train_labels, epochs=5)

test_loss, test_acc = model.evaluate(test_images, test_labels)
print("Test accuracy:", test_acc)


plt.imshow(test_images[0], cmap='gray')
plt.title(f"Prediction: {model.predict(test_images[:1]).argmax()}")
plt.show()
```
